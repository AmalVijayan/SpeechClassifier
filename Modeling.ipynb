```python
import tensorflow as tf
print("Tensorflow Version : ",tf.__version__)
import os
import re
import pandas as pd
```

## Loading The Data

```python
#Loading the training data
train = pd.read_excel("/Users/amalnair/Documents/Amal-WorkSpace/Speech-Classifier/SpeechClassification/Data_Train.xlsx")

#We will only use a fraction (20%) of the original dataset to train the model
train = train.sample(frac = 0.2)
```

```python
train.head()
```

## Initializing The Classifier

```python
#installing Simple Transformers
!pip install simpletransformers
```

```python
from simpletransformers.classification import ClassificationModel
```

```python
# Create a ClassificationModel
model = ClassificationModel('roberta', 'roberta-base', num_labels=4, use_cuda = False)
```

## Training The Model

```python
model.train_model(train)
```

## Evaluating The Model

```python
predictions, _ = model.predict(['Modi is the prime minister of India'])
```

```python
int(predictions)
```

```python
predictions2, _ = model.predict(['my phone is soo dumb and slow'])
```
